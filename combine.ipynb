```python
import pandas as pd
import glob
from tqdm import tqdm
```

```python
glob.glob("cat.csv")
```

```python
num = [i for i in range(1,18)]
weeks = [f'data/week{n}.csv' for n in num]
weeks
```

```python
total_df = []
for w in tqdm(weeks):
    total_df.append(pd.read_csv(w))
```

```python
data = pd.concat(total_df)
```

```python
data = data.reset_index(drop=True)
```

```python
data
```

```python
data.to_csv('data/all_tracking.csv',index=False)
```
